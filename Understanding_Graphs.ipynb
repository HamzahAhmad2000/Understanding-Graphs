```python
import networkx as nx
import matplotlib.pyplot as plt

# New meaningful knowledge graph triples with additional information
meaningful_triples = [
    # People
    ("Alice", "HasRole", "Engineer", {"age": 28}),
    ("Bob", "HasRole", "Manager", {"age": 35}),
    ("Charlie", "HasRole", "Artist", {"age": 30}),

    # Organizations
    ("TechCorp", "HostsEvent", "TechConference", {"location": "CityA", "est_year": 2000}),
    ("ArtStudio", "HostsEvent", "ArtExhibition", {"location": "CityB", "est_year": 2010}),

    # Events
    ("TechConference", "Attendee", "Alice", {"event_date": "2023-05-15"}),
    ("TechConference", "Attendee", "Bob", {"event_date": "2023-05-15"}),
    ("ArtExhibition", "Attendee", "Charlie", {"event_date": "2023-06-20"}),
    ("ArtExhibition", "Attendee", "Alice", {"event_date": "2023-06-20"}),
    # ... add more triples with features as needed
]

# Create a directed graph with additional node features
meaningful_graph = nx.DiGraph()
for triple in meaningful_triples:
    subject, predicate, obj, features = triple
    meaningful_graph.add_node(subject, **features)
    meaningful_graph.add_node(obj, **features)
    meaningful_graph.add_edge(subject, obj, relation=predicate)

# Visualize the graph
pos = nx.spring_layout(meaningful_graph)
nx.draw(meaningful_graph, pos, with_labels=True, font_weight='bold', node_color='skyblue', node_size=1000, edge_color='gray', linewidths=1, font_size=8)

plt.show()
```
